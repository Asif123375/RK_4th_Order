```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

def f(t,x,v):
  return v
def g(t,x,v):
  return -x

def RK_4th_Order(t0,x0,v0):
  h = 0.1
  x = []
  v = []
  t = []
  for i in range(1,210):
    k1 = h*f(t0,x0,v0)
    l1 = h*g(t0,x0,v0)
    k2 = h*f(t0 + h/2, x0 + k1/2, v0 + l1/2)
    l2 = h*g(t0 + h/2, x0 + k1/2, v0 + l1/2)
    k3 = h*f(t0 + h/2, x0 + k2/2, v0 + l2/2)
    l3 = h*g(t0 + h/2, x0 + k2/2, v0 + l2/2)
    k4 = h*f(t0 + h , x0 + k3, v0 + l3)
    l4 = h*g(t0 + h , x0 + k3, v0 + l3)
    xi = x0 + 1/6 *(k1 + 2*k2 + 2*k3 + k4)
    vi = v0 + 1/6 *(l1 + 2*l2 + 2*l3 + l4)
    x.append(x0)
    v.append(v0)
    t.append(t0)
    t0 += 0.1
    x0 = xi 
    v0 = vi
  return x, v, t

def plot_results():
  x ,v ,t = RK_4th_Order(0,1,0)
  plt.style.use('dark_background')
  plt.figure(figsize = (6,6),dpi = 150)
  plt.plot(t,x,'r-', label = "Position", linewidth = 2)
  plt.plot(t,v,'g', label = "Velocity", linewidth = 2)
  plt.xlabel('Time(t)')
  plt.ylabel("Results")
  plt.title('Solved SHO Using Rk_4th_Order')
  plt.legend()
plot_results()
```

```python
#Taking x,v,t
x,v,t = RK_4th_Order(0,1,0)
# list To array
x_array = np.array(x)
v_array = np.array(v)
t_array = np.array(t)
# Create DataFrame of Time,Position,Velocity
df = pd.DataFrame({"Time":t_array, "Position":x_array, "Velocity":v_array})
df.head(50)
```
